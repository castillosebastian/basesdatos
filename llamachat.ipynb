```python
import networkx as nx
import matplotlib.pyplot as plt

# Inicializar el gráfico
G = nx.DiGraph()

# Nodos
G.add_node("Usuario")
G.add_node("Base de Datos de Documentos")
G.add_node("Software")
G.add_node("Buscador")
G.add_node("Base de Datos Vectorizada")
G.add_node("FAISS")
G.add_node("Generador")
G.add_node("LLama2")

# Conexiones
G.add_edge("Usuario", "Software")
G.add_edge("Software", "Buscador")
G.add_edge("Software", "Generador")
G.add_edge("Buscador", "Base de Datos Vectorizada")
G.add_edge("Base de Datos Vectorizada", "FAISS")
G.add_edge("Generador", "LLama2")
G.add_edge("Buscador", "Base de Datos de Documentos")

# Dibujar con layout circular
pos = nx.circular_layout(G)
plt.figure(figsize=(10, 6))
nx.draw(G, pos, with_labels=True, node_size=3000, node_color="lightblue", font_size=12, font_weight='bold', width=2, edge_color="gray")

plt.title("Arquitectura del Chatbot")
plt.show()


```
